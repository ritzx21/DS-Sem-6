```python
import numpy as np
import matplotlib.pyplot as plt
```

MAX ABSOLUTE SCALING

```python
x = np.array([-3,-2,-1,0,1,2,3,4,5])
```

```python
max = np.abs(x).max()
```

```python
max
```

```python
max_abs_scaling = x/max
max_abs_scaling
```

```python
plt.plot(x,max_abs_scaling)
plt.title('Max Absolute Scaling')
plt.show()
```

MIN-MAX FEATURE SCALING

```python
max = np.max(x)
```

```python
max
```

```python
min = np.min(x)
```

```python
min
```

```python
minmax_feature = (x-min)/(max-min)
```

```python
minmax_feature
```

```python
plt.plot(minmax_feature)
plt.title("Min Max Feature Scaling")
plt.show()
```

Z-SCORE METHOD

```python
mean = np.mean(x)
mean
```

```python
std = np.std(x)
std
```

```python
z_score = (x-mean)/std
z_score
```

```python
plt.plot(x,z_score)
plt.title('Z_Score')
plt.show()
```
